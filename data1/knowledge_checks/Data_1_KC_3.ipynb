```python
import pandas as pd
import seaborn as sns
```

```python
housing_df = pd.read_csv('sample_data/california_housing_test.csv')
housing_df.head()
```

I'm using the same data set from KC 2, but wrote a few functions below to calculate various things from the data. Also notice the doc strings that come from PEP 257: https://peps.python.org/pep-0257/. These functions are also not perfect - they're somewhat inflexible because, in normalize() below, I have to have fields named longitude and latitude. If the dataframe doesn't have them, the function breaks. 

With this data as well, you could just as easily write the Pandas commands line by line rather than putting them in a function. Here's the important takeaway: If you find yourself copy-pasting a bunch of commands or code, try to put those lines into a function and then call the function when you need it. 

```python
def normalize(df):

  """
  This is a docstring. All good Python functions should have them per PEP257. 
  This will explain to future users what your function is supposed to do and 
  what the input is supposed to be. 

  Inputs:
  -------
  df : DataFrame object, a dataframe with columns "latitude" and "longitude"

  Returns:
  -------
  avg_coordinates : tuple of floats, a pair of the average latitude and long. 
  present in the dataframe
  """

  avg_long = sum(df.longitude) / df.shape[0] #longitude divided by number of rows 
  avg_lat = sum(df.latitude) / df.shape[0] #latitude divided by rows 
  avg_coordinates = (avg_long, avg_lat)
  return avg_coordinates

normalize(housing_df)
```

```python
def summarize(df):

  """
  A basic summary function for a given dataframe.

  Inputs:
  -------
  df : DataFrame object with columns "median_house_value" and "total_rooms". 
  
  Returns: 
  -------
  A seaborn histogram plot of the average housing prices.
  """

  avg_house = round(df.median_house_value.mean(),0)
  avg_rooms = round(df.total_rooms.mean(),0)
  print(f"The average median home price is ${avg_house} dollars")
  print(f"The average number of total rooms is {avg_rooms}")
  return sns.histplot(data=housing_df, x=housing_df.median_house_value); 
```

```python
summarize(housing_df); 
```

Here's an example of printing vs. returning a value in a function. 

```python
# This is NOT returning a value 

def print_addition(a,b):
  print(a+b)

print_addition(4,5)
```

```python
#This IS returning a value using the return statement

def return_addition(a,b):
  return a+b

result = return_addition(4,5)
result 
```

We can use the variable 'result' later like this: 

```python
result + 3
```

Oh no! When we try to add values to that print() function, it errors out. The print() actually returns NoneType, hence its inability to add 3 to None. 

```python
print_addition(4,5) + 3
```
